```python
def getAndParseURL(url):
    result = requests.get(url, headers = {"User-Agent": "Mozilla/5.0"})
    soup = bts(result.text, 'html.parser')
    return soup
```

```python
getAndParseURL("https://www.idefix.com/kategori/Kitap/Bilim/grupno=00052?ShowNotForSale=True")
```

```python
PAGE_NUM = []
def find_page_num():
    
    html = getAndParseURL("https://www.idefix.com/kategori/Kitap/Bilim/grupno=00052?ShowNotForSale=True")
    for i in range(1,11):
        PAGE_NUM.append("https://www.idefix.com/kategori/Kitap/Bilim/grupno=00052?ShowNotForSale=True"+str("&Page=")+str(i))
    return PAGE_NUM
find_page_num()
```

```python
html = getAndParseURL("https://www.idefix.com/kategori/Kitap/Bilim/grupno=00052?ShowNotForSale=True")

```

```python
for sonuc in html.findAll("div", {"class": "product-box-alternative product-box productListNewBox cart-product-box"}):
    print("https://www.idefix.com/"+sonuc.a["href"])
```

```python
ALL_PRODUCT = []
def all_product():
    
    for i in PAGE_NUM[::]:
        html = getAndParseURL(i)
        for sonuc in html.findAll("div", {"class": "product-box-alternative product-box productListNewBox cart-product-box"}):
            ALL_PRODUCT.append("https://www.idefix.com/"+sonuc.a["href"])
    return ALL_PRODUCT
all_product()
```

```python

result = []
for pr in ALL_PRODUCT[::]:
    
    prod = getAndParseURL(pr)
    #PRODUCT_UNQ = pr.split("-")[-1]
    Kitap_Adı = prod.find("h3", {"style": "margin-bottom: 10px !important; margin-top: 0px;"}).text.strip()
    Yazar_Adı = prod.findAll("a", {"class": "bold"})[1].text
    Fiyat = prod.find("div", {"class": "current-price"}).text.split(" ")[0]
    Sayfa_Sayısı = prod.findAll("a", {"class": "bold"})[4].text
    
    result.append([Kitap_Adı, Yazar_Adı, Fiyat, Sayfa_Sayısı])
    
column = ["Kitap_Adı", "Yazar_Adı", "Fiyat", "Sayfa_Sayısı"]
df = pd.DataFrame.from_records(result, columns=column)
```
